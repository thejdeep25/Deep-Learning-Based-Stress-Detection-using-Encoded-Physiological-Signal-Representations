```python
import pandas as pd
from pyts.image import MarkovTransitionField
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Dropout, Flatten, Input
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
import warnings
from collections import Counter
import gc
from tensorflow.keras import backend as K
from pyts.image import GramianAngularField

warnings.filterwarnings("ignore")

# Load dataset
df = pd.read_csv('/kaggle/input/cleaned/cleaned_dataset.csv')  # Replace with your path
df['DateTime'] = pd.to_datetime(df['DateTime'])
df = df.sort_values(['SubjectID', 'DateTime'])

# Preprocess features
feature_cols = ['AccX', 'AccY', 'AccZ', 'Temp', 'EDA', 'HeartRate', 'SpO2']
target_col = 'Label'

scaler = StandardScaler()
df[feature_cols] = scaler.fit_transform(df[feature_cols])

def create_windows(df, feature_cols, target_col, window_size=100, step=20):
    X = []
    y = []
    for subject_id, subject_df in df.groupby('SubjectID'):
        subject_data = subject_df.reset_index(drop=True)
        features = subject_data[feature_cols].values
        labels = subject_data[target_col].values
        for start in range(0, len(features) - window_size, step):
            end = start + window_size
            window_x = features[start:end]
            window_y = labels[start:end]
            label = Counter(window_y).most_common(1)[0][0]
            X.append(window_x)
            y.append(label)
    return np.array(X), np.array(y)

X, y = create_windows(df, feature_cols, target_col)

# Encode labels
le = LabelEncoder()
y_encoded = le.fit_transform(y)
y_cat = to_categorical(y_encoded)

# Convert time-series to images using MTF
def time_series_to_image_gaf(sample, image_size=64, method='difference'):
    gaf = GramianAngularField(image_size=image_size, method=method)
    img_list = []
    for i in range(sample.shape[1]):
        ts = sample[:, i]
        img = gaf.fit_transform(ts.reshape(1, -1))[0]
        img_list.append(img)
    img = np.stack(img_list, axis=-1)
    return img

image_data = np.array([time_series_to_image_gaf(seq) for seq in X])


# CNN model
def build_image_model(input_shape, num_classes):
    model = Sequential([
        Input(shape=input_shape),
        Conv2D(32, (3, 3), activation='relu', padding='same'),
        MaxPooling2D((2, 2)),
        Conv2D(64, (3, 3), activation='relu', padding='same'),
        MaxPooling2D((2, 2)),
        Conv2D(128, (3, 3), activation='relu', padding='same'),
        MaxPooling2D((2, 2)),
        Flatten(),
        Dense(128, activation='relu'),
        Dropout(0.5),
        Dense(num_classes, activation='softmax')
    ])
    model.compile(optimizer=Adam(1e-4), loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Evaluation
def evaluate_model(model, X_test, y_test, label_encoder, title="Model"):
    y_pred = model.predict(X_test)
    y_pred_classes = np.argmax(y_pred, axis=1)
    y_true_classes = np.argmax(y_test, axis=1)

    accuracy = accuracy_score(y_true_classes, y_pred_classes)
    precision = precision_score(y_true_classes, y_pred_classes, average='weighted')
    recall = recall_score(y_true_classes, y_pred_classes, average='weighted')
    f1 = f1_score(y_true_classes, y_pred_classes, average='weighted')

    cm = confusion_matrix(y_true_classes, y_pred_classes)
    specificity_per_class = []
    for i in range(len(cm)):
        tn = np.sum(np.delete(np.delete(cm, i, axis=0), i, axis=1))
        fp = np.sum(np.delete(cm, i, axis=0)[:, i])
        specificity = tn / (tn + fp) if (tn + fp) > 0 else 0
        specificity_per_class.append(specificity)
    weighted_specificity = np.average(specificity_per_class, weights=np.bincount(y_true_classes))

    print(f"\n{title}")
    print(f"Accuracy     : {accuracy:.4f}")
    print(f"Precision    : {precision:.4f}")
    print(f"Recall       : {recall:.4f}")
    print(f"F1 Score     : {f1:.4f}")
    print(f"Specificity  : {weighted_specificity:.4f}")

    return accuracy, precision, recall, f1, weighted_specificity


# LOSO Classification with memory clearing
results_loso = {}
subject_ids = df['SubjectID'].unique()

for test_subject in subject_ids:
    print(f"\n--- Evaluating on Subject {test_subject} ---")
    
    train_subjects = df[df['SubjectID'] != test_subject]
    test_subject_df = df[df['SubjectID'] == test_subject]
    
    X_train_loso, y_train_loso = create_windows(train_subjects, feature_cols, target_col)
    X_test_loso, y_test_loso = create_windows(test_subject_df, feature_cols, target_col)

    image_data_train = np.array([time_series_to_image_gaf(seq) for seq in X_train_loso])
    image_data_test = np.array([time_series_to_image_gaf(seq) for seq in X_test_loso])

    y_train_cat = to_categorical(le.transform(y_train_loso))
    y_test_cat = to_categorical(le.transform(y_test_loso))

    img_model_loso = build_image_model(image_data_train.shape[1:], y_cat.shape[1])
    img_model_loso.fit(image_data_train, y_train_cat,
                       validation_data=(image_data_test, y_test_cat),
                        epochs=15, batch_size=32)

    acc, precision, recall, f1, specificity = evaluate_model(
        img_model_loso, image_data_test, y_test_cat, le, title=f"Test Subject {test_subject}"
    )
    
    results_loso[test_subject] = {
        'Accuracy': acc,
        'Precision': precision,
        'Recall': recall,
        'F1': f1,
        'Specificity': specificity
    }
    # Save model for this subject
    model_path = f"/kaggle/working/loso_model_subject_{test_subject}.h5"
    img_model_loso.save(model_path)
    print(f"Saved model for Subject {test_subject} to {model_path}")
    
    # Clear memory
    K.clear_session()
    gc.collect()
    del img_model_loso, image_data_train, image_data_test, X_train_loso, X_test_loso, y_train_loso, y_test_loso, y_train_cat, y_test_cat

# Summary of results
print("\n=== LOSO Evaluation Summary ===")
print("Average Accuracy    :", np.mean([res['Accuracy'] for res in results_loso.values()]))
print("Average Precision   :", np.mean([res['Precision'] for res in results_loso.values()]))
print("Average Recall      :", np.mean([res['Recall'] for res in results_loso.values()]))
print("Average F1 Score    :", np.mean([res['F1'] for res in results_loso.values()]))
print("Average Specificity :", np.mean([res['Specificity'] for res in results_loso.values()]))
# Save LOSO Results
with open("/kaggle/working/loso_results.txt", "w") as f:
    f.write("=== LOSO Evaluation Results ===\n\n")
    for subject_id, metrics in results_loso.items():
        f.write(f"Test Subject {subject_id}:\n")
        for metric_name, metric_value in metrics.items():
            f.write(f"  {metric_name}: {metric_value:.4f}\n")
        f.write("\n")

    f.write("=== Average LOSO Metrics ===\n")
    f.write(f"Accuracy     : {np.mean([res['Accuracy'] for res in results_loso.values()]):.4f}\n")
    f.write(f"Precision    : {np.mean([res['Precision'] for res in results_loso.values()]):.4f}\n")
    f.write(f"Recall       : {np.mean([res['Recall'] for res in results_loso.values()]):.4f}\n")
    f.write(f"F1 Score     : {np.mean([res['F1'] for res in results_loso.values()]):.4f}\n")
    f.write(f"Specificity  : {np.mean([res['Specificity'] for res in results_loso.values()]):.4f}\n")
```
